# 4 PID Control

```python
import sys
sys.path.append('..')
from TCLab import TCLab, clock, pid
```

```python
a = TCLab()

c = pid()

Tsetpoint = 60
tfinal = 480
c.Kp = 10
c.Ki = 0.3

a.initplot()
a.start()

for t in clock(tfinal):
    a.Q1 = c.update(t,Tsetpoint,a.T1,a.Q1)
    a.updateplot()
    
a.stop()
```
