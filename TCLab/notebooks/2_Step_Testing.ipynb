## Step Test

```python
import sys
sys.path.append('..')
from TCLab import TCLab, clock

Qmax = 100
tfinal = 600
period = 200

log = []

a = TCLab()
a.start()
a.Q1 = Qmax
a.initplot()
for t in clock(tfinal):
    a.Q1 = 0 if (period%100 >= period/2) else Qmax
    log.append([t,a.Q1,a.T1,a.T2]) 
    a.updateplot()
    
a.stop()
```

```python
%matplotlib inline
import pandas as pd
df = pd.DataFrame(log,columns = ['Time','Q1','T1','T2'])
df = df.set_index('Time')
df.plot()
```
